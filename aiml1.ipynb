```python
year=range(2000,2008)
apples=[0.895,0.91,0.919,0.926,0.929,0.934,0.936,0.937]
oranges=[0.962,0.941,0.930,0.918,0.908,0.907,0.904,0.901]
```

```python
plt.plot(year,apples)
plt.plot(year,oranges)
plt.xlabel('year')
plt.ylabel('yield')
```

```python
plt.plot(year,apples,marker='o')
plt.plot(year,oranges,marker='x')

plt.ylabel('yield')
plt.title('crop yields in kanto')
plt.legend(['apples','oranges']);
```

```python
plt.plot(year,apples, 's-b')
plt.plot(year,oranges,'o--r')
plt.xlabel('years')
plt.ylabel('yield')

plt.title('crop yields in kanto')
plt.legend(['apples','oranges'])
```

```python
plt.plot(year,oranges,'or')
plt.title('yield of oranges')
```

```python
plt.figure(figsize=(4,4))
plt.plot(year,oranges,'or')
plt.title('yield of oranges')
```

```python
import matplotlib.pyplot as plt
x = [3,1,3,12,2,4,4]
y = [3,2,1,4,5,6,7]
plt.bar(x,y)
plt.title("Bar Chart")
plt.legend(["bar"])
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [3,1,3,12,2,4,4]
y = [3,2,1,4,5,6,7]
plt.scatter(x,y)
plt.legend("A")
plt.xlabel('x')
plt.ylabel('y')
plt.title("Scatter Chart")
plt.show
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4]
e = (0.1,0,0,0)
plt.pie(x, explode = e)
plt.title("Pie Chart")
plt.show()
```
